```python
#TASK 1
import numpy as np

A = np.random.rand(5)
B = np.random.rand(3,3)
C = np.random.rand(2,4)
D = np.random.rand(4,2)
E = np.random.rand(5,5)
F = np.random.rand(5)
G = np.random.rand(3,3)
H = np.random.rand(2,4)
I = np.random.rand(4,2)
J = np.random.rand(5,5)

# Single Array
np.sqrt(A)
np.square(A)
np.log10(F)
np.tan(J)
np.cumsum(G)
np.mean(C) 
np.std(G) 
np.var(D)
np.copysign(C,-3)
np.rint(E)  
np.arccos(I)
np.diag(G)

# Same size Array
np.maximum(A,F) 
np.subtract(D,I) 
np.add(B,G)
np.minimum(B,G)
np.power(B,G) 

# Different size Array
np.multiply(D,C.T) 
np.divide(I,H.T) 

print('Single Array')

print("1"),print(np.sqrt(A))
print('\n--------\n')

print("2"),print(np.square(A))
print('\n--------\n')

print("3"),print(np.log10(F))
print('\n--------\n')

print("4"),print(np.tan(J))
print('\n--------\n')

print("5"),print(np.cumsum(G))
print('\n--------\n')

print("6"),print(np.mean(C))
print('\n--------\n')

print("7"),print(np.std(G))
print('\n--------\n')

print("8"),print(np.var(D))
print('\n--------\n')

print("9"),print(np.copysign(C,-3))
print('\n--------\n')

print("10"),print(np.rint(E))
print('\n--------\n')

print("11"),print(np.arccos(I))
print('\n--------\n')

print("12"),print(np.diag(G))
print('\n--------\n')

print('Same Size Array')

print("13"),print(np.maximum(A,F))
print('\n--------\n')

print("14"),print(np.subtract(D,I))
print('\n--------\n')

print("15"),print(np.add(B,G))
print('\n--------\n')

print("16"),print(np.minimum(B,G))
print('\n--------\n')

print("17"),print(np.power(B,G))
print('\n--------\n')

print('Different Size Array')

print("18"),print(np.dot(C,D))
print('\n--------\n')

print("19"),print(np.multiply(D,C.T))
print('\n--------\n')

print("20"),print(np.divide(I,H.T))
```
